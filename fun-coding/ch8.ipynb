```python
df = pd.read_csv('seoul_apartment.csv')
average_price = df['거래금액(만원)'].mean()
print("평균 매매가격:", average_price)
```

```python
grouped = df.groupby('시군구')
average_prices = grouped['거래금액(만원)'].mean()
print("지역별 평균 매매가격:")
print(average_prices)
```

```python
total_prices = df.groupby('시군구')['거래금액(만원)'].sum()
print("지역별 총 거래금액:")
print(total_prices)
```

```python
max_area = df.groupby('단지명')['전용면적(㎡)'].max()
print("단지별 최대 전용면적(㎡):")
print(max_area)
```

```python
yearly_transactions = df.groupby('건축년도')['NO'].count()
print("건축년도별 거래 횟수:")
print(yearly_transactions)
```

```python
transactions = df.groupby('계약년월')['NO'].count()
print("계약년월별 거래 횟수:")
print(transactions)
```

```python
import pandas as pd
df = pd.read_csv('seoul_apartment.csv')
print(df['거래금액(만원)'].min())
print(df['거래금액(만원)'].max())
print(df['거래금액(만원)'].mean())
print(df['거래금액(만원)'].median())
```

```python
average_prices = df.groupby('시군구')['거래금액(만원)'].mean()
top_5 = average_prices.sort_values(ascending=False).head(5)
print("평균 거래금액 상위 5개 지역:")
print(top_5)
```

```python
gangnam_sinsa = df[df['시군구'] == "서울특별시 강남구 신사동"]
print(gangnam_sinsa['거래금액(만원)'].mean())
print(gangnam_sinsa['거래금액(만원)'].max())
```
